### **1. Grafos Clásicos**
NetworkX puede generar una serie de grafos clásicos utilizados comúnmente en la teoría de grafos y en la investigación de redes.

**1.1 Grafo Completo**
Un grafo completo es un grafo no dirigido en el que cada par de nodos distintos está conectado por un enlace único.



```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.complete_graph(100)
nx.draw(G, with_labels=True)
plt.title("Grafo Completo")
plt.show()
```

**1.2 Grafo de Petersen**
El grafo de Petersen es un grafo no dirigido con 10 vértices y 15 aristas. Es un ejemplo clásico utilizado en la teoría de grafos.

```python
G = nx.petersen_graph()
nx.draw_shell(G, with_labels=True)
plt.title("Grafo de Petersen")
plt.show()
```

### **2.1 Zachary's Karate Club**
Este es un grafo que representa la red social del club de karate en una universidad de EE. UU., descrito por Wayne W. Zachary en 1977. Es una red bien conocida y a menudo se utiliza como ejemplo en la literatura de redes.

## **3. Personalización de Grafos**
Una vez que hemos generado o cargado un grafo preexistente, podemos personalizar su apariencia y extraer información valiosa.

### **3.1 Colorear nodos basados en atributos**
Supongamos que queremos colorear los nodos del Grafo del Club de Karate basados en el club al que se unieron después de una división

```python
club_assignment = nx.get_node_attributes(G, 'club')
colors = ["red" if club_assignment[node] == "Mr. Hi" else "blue" for node in G.nodes()]

nx.draw_spring(G, with_labels=True, node_color=colors)
plt.title("Karate Club Split")
plt.show()
```

#Tipos de Archivos

***- Lista de aristas (Edge List):***

En NetworkX, una "lista de aristas" (Edge List) es una forma común de representar un grafo en términos de sus conexiones entre nodos. Cada elemento en la lista de aristas representa una arista o conexión entre dos nodos en el grafo. Por lo general, se representa como una lista de tuplas donde cada tupla contiene dos nodos que están conectados por una arista. Un ejemplo de cómo se vería una lista de aristas en NetworkX:

```python
edge_list = [(1, 2), (2, 3), (1, 4), (4, 5)]
```

  **Ventajas de la lista de aristas en NetworkX:**

  **Sencillez:** La lista de aristas es una forma simple y directa de representar las conexiones entre nodos en un grafo. Es fácil de entender y de implementar.

  **Eficiencia en almacenamiento:** Si el grafo es disperso (tiene relativamente pocas aristas en comparación con el número total de nodos posibles), la lista de aristas puede ser una forma eficiente de almacenar el grafo, ya que solo almacena información sobre las aristas presentes.

  **Facilidad de construcción:** Puede ser más fácil construir un grafo desde una lista de aristas si estás trabajando con datos tabulares o estructurados donde las aristas se representan como pares de nodos.

  **Desventajas de la lista de aristas en NetworkX:**

  **Ineficiencia en búsqueda:** Si necesitas buscar si existe una arista específica entre dos nodos, la lista de aristas puede ser ineficiente, ya que requerirá recorrer la lista completa para buscar la arista deseada.

  **Ineficiencia en la obtención de vecinos:** Si necesitas obtener los vecinos de un nodo específico, la lista de aristas no es la estructura más eficiente, ya que requerirá buscar todas las aristas que involucran al nodo.

  **Almacenamiento innecesario:** En grafos densos (con muchas aristas), la lista de aristas puede ocupar más espacio de almacenamiento que otras estructuras más eficientes como la matriz de adyacencia o la lista de adyacencia.

***- Matriz de adyacencia:***

La matriz de adyacencia en NetworkX es una forma de representar un grafo utilizando una matriz bidimensional, donde las filas y columnas representan nodos y los valores en la matriz indican si existe una arista entre los nodos correspondientes. A continuación, te mostraré cómo se ve una matriz de adyacencia en NetworkX y luego discutiremos sus ventajas y desventajas.

```python
 matriz = np.array([
    [0, 1, 1, 0], # A
    [1, 0, 0, 0], # B
    [1, 0, 0, 0], # C
    [0, 0, 0, 0]  # D
  ])
```

**Ventajas de la matriz de adyacencia en NetworkX:**

  **Eficiencia en la verificación de la existencia de aristas:** Es eficiente para verificar si existe una arista entre dos nodos específicos, ya que simplemente se busca el valor en la matriz en lugar de recorrer una lista de aristas.

 **Eficiencia en la obtención de vecinos:** Es eficiente para encontrar los vecinos de un nodo específico, ya que solo se necesita examinar una fila o columna en la matriz.

  **Fácil visualización:** La matriz de adyacencia proporciona una representación visual clara de la estructura del grafo y es especialmente útil para grafos pequeños.

  **Desventajas de la matriz de adyacencia en NetworkX:**

  **Ineficiencia en el almacenamiento:** Puede ocupar una cantidad significativa de espacio de almacenamiento, especialmente para grafos grandes o densos, ya que debe almacenar información sobre todas las posibles aristas, incluso si no están presentes en el grafo.

  **Ineficiencia en la actualización:** Si el grafo es dinámico y cambia con el tiempo, actualizar la matriz de adyacencia para reflejar esos cambios puede ser costoso en términos de tiempo y espacio.

  **No adecuado para grafos ponderados:** La matriz de adyacencia es menos adecuada para representar grafos ponderados, ya que solo puede almacenar información binaria sobre la presencia o ausencia de aristas.

***- GEXF (Graph Exchange XML Format):***

GEXF (Graph Exchange XML Format) es un formato estándar para representar grafos en forma de documentos XML. NetworkX, una biblioteca de Python para el análisis de redes, admite la lectura y escritura de grafos en formato GEXF. A continuación, te mostraré cómo se ve un ejemplo de archivo GEXF y las ventajas y desventajas de su uso en NetworkX, utilizando ejemplos en Python.

**Aspecto de un archivo GEXF en Python:**

Para crear y trabajar con archivos GEXF en NetworkX, primero debes cargar un grafo desde un archivo existente o crear uno desde cero y luego guardar el grafo en formato GEXF. Aquí tienes un ejemplo de cómo se vería esto en Python:

```python
import networkx as nx

# Crear un grafo
G = nx.Graph()
G.add_node(1)
G.add_node(2)
G.add_edge(1, 2)

# Guardar el grafo en formato GEXF
nx.write_gexf(G, "mi_grafo.gexf")

# Cargar un grafo desde un archivo GEXF
H = nx.read_gexf("mi_grafo.gexf")
```

**Ventajas de GEXF en NetworkX:**

**Interoperabilidad:** GEXF es un formato estándar ampliamente aceptado, lo que facilita compartir datos de grafos con otras aplicaciones y bibliotecas que admiten GEXF.

**Flexibilidad:** GEXF admite atributos en nodos y aristas, lo que te permite agregar metadatos o información adicional a tu grafo de manera estructurada.

**Legibilidad:** Los archivos GEXF son legibles por humanos porque utilizan la sintaxis XML, lo que facilita la inspección y edición manual cuando sea necesario.

**Desventajas de GEXF en NetworkX:**

**Tamaño del archivo:** Los archivos GEXF pueden volverse relativamente grandes para grafos grandes o densos, lo que puede requerir más espacio de almacenamiento y ancho de banda para transferirlos.

**Complejidad:** La estructura XML puede ser compleja para grafos muy simples, lo que puede hacer que la representación sea un poco verbosa en casos sencillos.

**Rendimiento:** La lectura y escritura de archivos GEXF pueden ser más lentas que otros formatos de archivo más compactos, especialmente para grafos muy grandes.


***- GraphML:***

GraphML es otro formato estándar utilizado para representar grafos, al igual que GEXF, pero utiliza una estructura basada en XML para describir los nodos, aristas y atributos de un grafo. NetworkX, una biblioteca de Python para análisis de redes, admite la lectura y escritura de grafos en formato GraphML.
