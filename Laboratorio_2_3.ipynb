## 2. Análisis de Redes de Jugadores en un Torneo Deportivo

### Descripción:
A partir de los partidos jugados en un torneo deportivo (seleccionar uno de la lista proporcionada), se construirá una red donde cada jugador es un nodo y cada partido jugado entre dos jugadores es una arista.

### Objetivos:
* Analizar la competitividad y la diversidad de los enfrentamientos.
* Identificar jugadores clave y rivalidades destacadas.

### Tareas:

1. **Construcción del Grafo:**
   - Construye el grafo basándose en la información de torneos que este disponible.

2. **Distancia entre Nodos:**
   - Utiliza la métrica de "Distancia entre nodos" para entender las "rivalidades" entre jugadores.

3. **Centro de la Gráfica:**
   - Encuentra el "Centro de la gráfica" para identificar los jugadores que tuvieron más partidos y, por ende, pueden considerarse como figuras centrales del torneo.

4. **Periferia:**
   - La "Periferia" podría indicar jugadores que tuvieron menos partidos o que fueron menos centrales en el torneo.

5. **Excentricidad:**
   - Analiza la "Excentricidad" para identificar jugadores que tuvieron una variedad más amplia de oponentes.

- **Tenis:**
   - Ejemplo de Torneo: *Grand Slam* (Australian Open, Roland Garros, Wimbledon, US Open).
   - Fuentes de Datos: *ATP y WTA*.

- **Badminton:**
   - Estructura similar al tenis pero con menor cobertura en algunas áreas.
   
- **Squash:**
   - Menos popular que el tenis, pero con torneos y datos similares, especialmente para eventos principales.

- **Ajedrez:**
   - Ejemplo de Torneo: *Torneo de Candidatos, Campeonato Mundial*.
   - Nota: Aunque no es un deporte físico, tiene estructura de torneo y extensas bases de datos.

- **Boxeo:**
   - Estructura de torneo variable. Datos disponibles para combates en principales divisiones y categorías de peso.

- **Golf:**
   - Ejemplo de Torneo: *Majors* (Masters, US Open, The Open Championship, PGA Championship).
   
- **Mesa de billar o Pool:**
   - Algunos torneos presentan amplia participación y datos sobre enfrentamientos.
