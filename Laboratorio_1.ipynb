# **Introducción a Redes Sociales con NetworkX**
# 1. Creación de una Red Simple
Antes de empezar con la manipulación y visualización de redes, es esencial entender los conceptos fundamentales. Una red se compone de nodos y aristas. En este ejercicio, crearemos una red simple para entender estos componentes y cómo se relacionan entre sí.

## 1.1. Importar bibliotecas
Para este ejercicio, utilizaremos dos bibliotecas:

-  **networkx**: Una biblioteca especializada en la creación, manipulación y estudio de la estructura, dinámica y funciones de redes complejas.
-  **matplotlib**: Una bibliotecaampliamente utilizada para producir visualizaciones estáticas, animadas e interactivas en Python.



```python
import networkx as nx
import matplotlib.pyplot as plt
```

# 1.2. Crear y visualizar una red simple
Ahora que hemos importado las bibliotecas necesarias, vamos a:

1. Crear un grafo vacío.

2. Añadir nodos a este grafo.
3. Añadir aristas entre estos nodos.
4. Visualizar la red resultante.

```python
# Crear un grafo vacío
G = nx.Graph()

# Añadir nodos
# Los nodos son entidades en la red. Por ejemplo, en una red social, los nodos pueden representar personas.
G.add_node("A")
G.add_node("B")
G.add_node("C")
G.add_node("D")

# Añadir aristas
# Las aristas representan relaciones entre nodos. En una red social, una arista podría representar una amistad.
G.add_edge("A", "B")
G.add_edge("B", "C")
G.add_edge("A", "C")
G.add_edge("C", "D")

# Dibujar el grafo
# La visualización nos ayuda a entender la estructura de la red.
nx.draw(G, with_labels=True, node_color='skyblue', node_size=1000)
plt.show()
```

# **2. Métricas Básicas de la Red**
Una vez que tenemos una red, es útil entender algunas métricas básicas.

2.1. Número de nodos y aristas
Este código te ayudará a entender el tamaño de tu red.

```python
print("Número de nodos:", G.number_of_nodes())
print("Número de aristas:", G.number_of_edges())
```

# 2.2. Vecinos de un nodo
 Los vecinos de un nodo son aquellos conectados directamente a él. Esta métrica es esencial para entender las interacciones directas de una entidad en la red.

```python
print("Vecinos del nodo C:", list(G.neighbors("C")))
```

# **3. Experimentación con Visualizaciones**
La visualización de redes puede variar en función de las necesidades y de lo que se quiera destacar. Experimentaremos con diferentes formas de visualizar nuestra red simple.

# 3.1. Visualización Circular
Esta visualización organiza los nodos en un círculo, lo que puede ser útil para entender ciclos o patrones circulares en una red.

```python
nx.draw_circular(G, with_labels=True, node_color='skyblue', node_size=1000)
plt.show()
```

# 3.2. Visualización Kamada-Kawai
El layout Kamada-Kawai intenta posicionar los nodos de una manera que reduzca la cantidad de cruces entre aristas y destaque la estructura de la red.

```python
nx.draw_kamada_kawai(G, with_labels=True, node_color='skyblue', node_size=1000)
plt.show()
```
